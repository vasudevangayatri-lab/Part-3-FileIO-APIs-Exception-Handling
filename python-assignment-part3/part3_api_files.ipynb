```python
# Install the requests library and import needed components

import requests
import json

BASE_URL = "https://dummyjson.com/products"

# GET requests 

def safe_get(url):
    
    try:
        r = requests.get(url, timeout=10)
        r.raise_for_status()
        return r.json()
    except requests.exceptions.ConnectionError:
        print(f"  - Connection Error for: {url}")
        return None
    except requests.exceptions.Timeout:
        print(f"  - Timeout for: {url}")
        return None
    except requests.exceptions.HTTPError as e:
        print(f"  - HTTP Error for {url}: {e}")
        return None

# STEP 1 — Fetch and display 20 products

print("=" * 65)
print(" STEP 1 — Fetch 20 Products")
print("=" * 65)

data = safe_get(f"{BASE_URL}?limit=20")

if data:
    products = data["products"]
    print(f"\n  - {len(products)} products fetched from DummyJSON API\n")

    print(f"{'ID':<4}| {'Title':<32}| {'Category':<15}| {'Price':<10}| {'Rating'}")
    print("-" * 4 + "+" + "-" * 33 + "+" + "-" * 16 + "+" + "-" * 11 + "+" + "-" * 7)

    for p in products:
        title = p["title"] if len(p["title"]) <= 31 else p["title"][:28] + "..."
        print(f"{p['id']:<4}| {title:<32}| {p['category']:<15}| ${p['price']:<9.2f}| {p['rating']}")
else:
    products = []
    print("  Step 1 skipped due to API error.")


# STEP 2 — Filter rating >= 4.5, sort by price descending

print("\n" + "=" * 65)
print(" STEP 2 — Filter (rating ≥ 4.5) & Sort (price high to low)")
print("=" * 65)

if products:
    high_rated     = [p for p in products if p["rating"] >= 4.5]
    filtered_sorted = sorted(high_rated, key=lambda p: p["price"], reverse=True)

    print(f"\n  {len(high_rated)} of {len(products)} products have rating ≥ 4.5\n")
    print(f"{'Rank':<5}| {'Title':<32}| {'Category':<15}| {'Price':<10}| {'Rating'}")
    print("-" * 5 + "+" + "-" * 33 + "+" + "-" * 16 + "+" + "-" * 11 + "+" + "-" * 7)

    for rank, p in enumerate(filtered_sorted, start=1):
        title = p["title"] if len(p["title"]) <= 31 else p["title"][:28] + "..."
        print(f"{rank:<5}| {title:<32}| {p['category']:<15}| ${p['price']:<9.2f}| {p['rating']}")
else:
    print("  Step 2 skipped — no product data available.")

# STEP 3 — Fetch by category (all laptops)

print("\n" + "=" * 65)
print(" STEP 3 — Search by Category: 'laptops'")
print("=" * 65)

laptops_data = safe_get(f"{BASE_URL}/category/laptops")

if laptops_data:
    laptops = laptops_data["products"]
    print(f"\n  - {len(laptops)} laptop(s) found\n")

    print(f"{'#':<4} {'Product Name':<42} {'Price':>8}")
    print("-" * 58)
    for i, lp in enumerate(laptops, start=1):
        print(f"{i:<4} {lp['title']:<42} ${lp['price']:>7.2f}")

    prices   = [lp["price"] for lp in laptops]
    print("-" * 58)
    print(f"  Average price: ${sum(prices)/len(prices):.2f}")
else:
    print("  Step 3 skipped due to API error.")

# STEP 4 —  POST a new product Request

print("\n" + "=" * 65)
print(" STEP 4 — POST Request: Create a New Product")
print("=" * 65)

new_product = {
    "title":       "My Custom Product",
    "price":       999,
    "category":    "electronics",
    "description": "A product I created via API",
}

print("\n  Payload sent:")
print(json.dumps(new_product, indent=4))

try:
    r_post = requests.post(f"{BASE_URL}/add", json=new_product, timeout=10)
    r_post.raise_for_status()
    created = r_post.json()
    print(f"\n  - POST successful! Status: {r_post.status_code}")
    print("\n  Server response:")
    print(json.dumps(created, indent=4))
except Exception as e:
    print(f"  - POST failed: {e}")

print("\n" + "=" * 65)
print(" --> Task 2 Complete!")
print("=" * 65)
```

### Part A — Guarded Calculator

```python
def safe_divide(a, b):
    
    try:
        result = a / b
        return result

    except ZeroDivisionError:
        # Runs when divide by zero
        return "Error: Cannot divide by zero"

    except TypeError:
        # Runs when a or b is a non-numeric type 
        return "Error: Invalid input types"

# Task specific Required 3 tests

print("=" * 50)
print(" Part A — safe_divide tests")
print("=" * 50)

# Test 1: Normal division 
result1 = safe_divide(10, 2)
print(f"  safe_divide(10, 2)      -> {result1}")

# Test 2: Divide by zero 
result2 = safe_divide(10, 0)
print(f"  safe_divide(10, 0)      -> {result2}")

# Test 3: Wrong type of input
result3 = safe_divide("ten", 2)
print(f"  safe_divide('ten', 2)   -> {result3}")
```

```python
# read_file_safe: FileNotFoundError, and finally block

def read_file_safe(filename):
    
    try:
        with open(filename, "r", encoding="utf-8") as f:
            content = f.read()    
        print(f"  File '{filename}' read successfully "
              f"({len(content)} characters, "
              f"{content.count(chr(10))} lines).")
        return content            

    except FileNotFoundError:
        print(f"  Error: File '{filename}' not found.")
        return None               

    finally:
        print("  File operation attempt complete.")

# Test 1: Existing file 

print("=" * 50)
print(" Test 1 — Read an EXISTING file")
print("=" * 50)
result_good = read_file_safe("python_notes.txt")

if result_good is not None:
    print()
    print("  First 3 lines of content:")
    for i, line in enumerate(result_good.strip().split("\n")[:3], start=1):
        print(f"    {i}. {line}")

# Test 2: NOT EXISTing file

print()
print("=" * 50)
print(" Test 2 — Read a MISSING file (ghost_file.txt)")
print("=" * 50)
result_bad = read_file_safe("ghost_file.txt")

print()
print(f"  Return value when file missing: {result_bad}")
print("  (None means we can safely check: if result is None -> handle failure)")
```

### Part C — Robust API Calls

```python
# distinction between python and HTTP level error
# Python-level exceptions raised BEFORE a response exists
# HTTP-level errors - server was reached, and sent back a response

import subprocess, sys
subprocess.check_call([sys.executable, "-m", "pip", "install", "requests", "-q"])
import requests

print("requests exception class hierarchy:")
print()
print("  requests.exceptions.RequestException   <- base class (all requests errors)")
print("  ├── ConnectionError                    <- NO internet / DNS failure")
print("  ├── Timeout                            <- server too slow (> timeout seconds)")
print("  └── HTTPError                          <- server reached but bad status code")
print()
print("  Separate from Python exceptions:")
print("  ├── FileNotFoundError  (built-in)")
print("  ├── ZeroDivisionError  (built-in)")
print("  └── ValueError         (built-in) ← used in Part D")
print()
print("-- Import complete. Base URL defined.")

BASE_URL = "https://dummyjson.com/products"
```

```python
# robust_get and robust_post: full exception-handling wrappers

def robust_get(url, timeout=5):
    
    try:
        response = requests.get(url, timeout=timeout)
        response.raise_for_status()
        return response.json()

    except requests.exceptions.ConnectionError:
        print("  Connection failed. Please check your internet.")
        return None

    except requests.exceptions.Timeout:
        # Server took longer than 'timeout' seconds to respond
        print("  Request timed out. Try again later.")
        return None

    except requests.exceptions.HTTPError as e:
        raise   

    except Exception as e:
        print(f"  Unexpected error: {e}")
        return None

def robust_post(url, payload, timeout=5):
    
    try:
        response = requests.post(url, json=payload, timeout=timeout)
        response.raise_for_status()
        return response.json()

    except requests.exceptions.ConnectionError:
        print("  Connection failed. Please check your internet.")
        return None

    except requests.exceptions.Timeout:
        print("  Request timed out. Try again later.")
        return None

    except requests.exceptions.HTTPError as e:
        print(f"  HTTP Error: {e}")
        return None

    except Exception as e:
        print(f"  Unexpected error: {e}")
        return None

print("robust_get() and robust_post() functions defined.")
```

### Part D — Input Validation Loop

```python
print("Part D concepts — input validation flow:")  
print()

test_inputs = ["abc", "0", "150", "-5", "3.14", "42", "quit"]

print(f"  {'Input':<10} {'int() result':<22} {'Range OK?':<12} {'Action'}")
print("  " + "-" * 62)

for raw in test_inputs:
    if raw.lower() == "quit":
        print(f"  {raw:<10} {'—':<22} {'—':<12} Exit loop")
        continue
    try:
        pid = int(raw)
        in_range = "Yes" if 1 <= pid <= 100 else "No"
        action   = "Call API" if in_range == "Yes" else "Warn + retry"
        print(f"  {raw:<10} {str(pid):<22} {in_range:<12} {action}")
    except ValueError:
        print(f"  {raw:<10} {'ValueError':<22} {'—':<12} Warn + retry")

print()
print("  This is exactly how the loop makes its decisions.")
```

```python
# Interactive product lookup loop 
# (requires internet + keyboard input)

print("─" * 55)
print(" Product Lookup — type a product ID or 'quit' to stop")
print("─" * 55)

while True:    

    user_input = input("\nEnter a product ID to look up (1–100), or 'quit' to exit: ").strip()
    if user_input.lower() == "quit":
        print("Exiting product lookup. Goodbye!")
        break
    try:
        product_id = int(user_input)   
    except ValueError:
        print(f"  - Warning: '{user_input}' is not a valid integer. Please enter a number.")
        continue    

    # Validate: range 1–100 
    if not (1 <= product_id <= 100):
        print(f"  - Warning: {product_id} is out of range. Please enter a number between 1 and 100.")
        continue   

    # Make the API call 
    url = f"{BASE_URL}/{product_id}"    

    try:
        response = requests.get(url, timeout=5)
        if response.status_code == 200:
            product = response.json()
            print(f"  -- Product #{product_id} found:")
            print(f"     Title    : {product.get('title', 'N/A')}")
            print(f"     Price    : ${product.get('price', 'N/A')}")
            print(f"     Category : {product.get('category', 'N/A')}")
            print(f"     Rating   : {product.get('rating', 'N/A')}")

        elif response.status_code == 404:
            # Server was reached but this product ID doesn't exist
            print(f"  -- Product not found. (ID {product_id} returned 404)")
        else:
            # Any other status code (500, 503, etc.)
            print(f"  -- Unexpected status code: {response.status_code}")

    except requests.exceptions.ConnectionError:
        print("  Connection failed. Please check your internet.")

    except requests.exceptions.Timeout:
        print("  Request timed out. Try again later.")

    except Exception as e:
        print(f"  Unexpected error: {e}")
```

### Task 4 — Logging to File

```python

# Setup log_error(), fetch_products(BAD_URL), lookup_product(999)

import requests
from datetime import datetime

LOG_FILE = "error_log.txt"
BASE_URL = "https://dummyjson.com/products"
BAD_URL  = "https://this-host-does-not-exist-xyz.com/api"

def log_error(function_name, error_type, message):
    """Write one timestamped error entry to error_log.txt using append mode."""
    timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    entry     = f"[{timestamp}] ERROR in {function_name}: {error_type} — {message}\n"
    with open(LOG_FILE, "a", encoding="utf-8") as f:   # 'a' never overwrites
        f.write(entry)
    print(f"  Logged → {entry.strip()}")

#  Routine for fetch_products 

def fetch_products(url):
    try:
        r = requests.get(url, timeout=5)
        r.raise_for_status()
        return r.json()
    except requests.exceptions.ConnectionError as e:
        log_error("fetch_products", "ConnectionError", str(e)[:80])
        return None
    except requests.exceptions.Timeout:
        log_error("fetch_products", "Timeout", "Request exceeded 5 second timeout")
        return None
    except requests.exceptions.HTTPError as e:
        log_error("fetch_products", "HTTPError", str(e)[:80])
        return None
    except Exception as e:
        log_error("fetch_products", type(e).__name__, str(e)[:80])
        return None

# Routine for lookup_product 

def lookup_product(product_id):
    """GET request — logs 404 via status_code check (not try/except)."""
    try:
        r = requests.get(f"{BASE_URL}/{product_id}", timeout=5)
        # Manual status check — NOT raise_for_status()
        if r.status_code == 200:
            return r.json()
        elif r.status_code == 404:
            # 404 is an HTTP response, not a Python exception — use an if block
            log_error("lookup_product", "HTTPError", f"404 Not Found for product ID {product_id}")
            return None
        else:
            log_error("lookup_product", "HTTPError", f"{r.status_code} Unexpected status for ID {product_id}")
            return None
    except requests.exceptions.ConnectionError as e:
        log_error("lookup_product", "ConnectionError", str(e)[:80])
        return None
    except requests.exceptions.Timeout:
        log_error("lookup_product", "Timeout", "Request exceeded 5 second timeout")
        return None
    except Exception as e:
        log_error("lookup_product", type(e).__name__, str(e)[:80])
        return None

print("=" * 60)
print(" Task 4 — Logging to File — Full Run")
print("=" * 60)

# Trigger 1: ConnectionError
print("\n── Trigger 1: ConnectionError (unreachable URL) ──")
fetch_products(BAD_URL)

# Trigger 2: HTTP 404
print("\n── Trigger 2: HTTP 404 (product ID 999) ──")
lookup_product(999)

# Bonus: confirm a valid lookup works without logging
print("\n── Bonus: valid lookup (product ID 1) ──")
p = lookup_product(1)
if p:
    print(f"  Success: {p.get('title')} — ${p.get('price')}")
    print("  (No log entry — 200 responses are not errors)")

# Display the log file
print()
print("=" * 60)
print(f" Contents of '{LOG_FILE}'")
print("=" * 60)
try:
    with open(LOG_FILE, "r", encoding="utf-8") as f:
        content = f.read()
    lines = content.splitlines()
    for i, line in enumerate(lines, start=1):
        print(f"  {i:>3}. {line}")
    print("=" * 60)
    print(f"  Total entries logged: {len(lines)}")
except FileNotFoundError:
    print(f"  '{LOG_FILE}' not found.")

print()
print("Task 4 Complete!")
```
